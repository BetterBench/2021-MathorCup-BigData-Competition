# 4 查看分布

```python
# 所有字段的分布
dist_cols = 4
dist_rows = len(test.columns)
plt.figure(figsize=(4*dist_cols,4*dist_rows))
i = 1
for col in column[:8]:
    if col == '价格' or col == '国别' or col == '匿名特征3':
        continue
    ax = plt.subplot(dist_rows,dist_cols,i)
    ax = sns.kdeplot(train[col],color='Blue',shade= True)
    ax  = sns.kdeplot(test[col],color='Red',shade=True)
    ax.set_xlabel(col,fontsize=15)
    ax = ax.legend(['训练集','测试集'],prop = {'size':15})
    i+=1

plt.savefig('img/Q1-img/所有字段分布1.png',dpi=300)
plt.show()
```

```python
dist_cols = 4
dist_rows = len(test.columns)
plt.figure(figsize=(4*dist_cols, 4*dist_rows))
i = 1
for col in column[8:]:
    if col == '价格' or col == '国别' or col == '匿名特征3':
        continue
    ax2 = plt.subplot(dist_rows,dist_cols,i)
    ax2 = sns.kdeplot(train[col],color='Blue',shade= True)
    ax2  = sns.kdeplot(test[col],color='Red',shade=True)
    ax2.set_xlabel(col)
    ax2 = ax2.legend(['训练集','测试集'],prop = {'size':15})
    i+=1
plt.savefig('img/Q1-img/所有字段分布2.png',dpi=300)
plt.show()
```

结论：如果出现某个特征数据集和测试集分布不一致，选择删除该特征。通过以上KDE分布图分析，可以看出数据集和测试集的分布一致，选择保留这些特征

```python
plt.figure(figsize=(20, 6))
y = train[train['价格']<700]['价格']
plt.subplot(121)
plt.title('johnsonsu分布拟合',fontsize=20)
plt.xlabel('价格',fontsize=15)
sns.distplot(y, kde=False, fit=st.johnsonsu, color='Red')

plt.subplot(121)
y2 = train[train['价格'] < 200]['价格']
plt.subplot(122)
st.probplot(y2, dist="norm", plot=plt)
plt.title('价格 Q-Q图',fontsize=20)
plt.xlabel('X分位数',fontsize=15)
plt.ylabel('Y分位数',fontsize=15)
plt.savefig('img/Q1-img/价格未转换前的分布.png',dpi=300)
plt.show()
```

数据分布并不符合正态分布，是右偏数据，回归中对数据分布较为敏感，如果不符合正态分布需要进行数据转换成近似正态分布

```python
plt.figure(figsize=(20, 6))
y = np.log1p(train[train['价格']<700]['价格'])
plt.subplot(121)
plt.title('johnsonsu分布拟合',fontsize=20)
plt.xlabel('价格',fontsize=15)
sns.distplot(y, kde=False, fit=st.johnsonsu, color='Red')

plt.subplot(121)
y2 = np.log1p(train[train['价格'] < 700]['价格'])
plt.subplot(122)
st.probplot(y2, dist="norm", plot=plt)
plt.title('价格 Q-Q图',fontsize=20)
plt.xlabel('X分位数',fontsize=15)
plt.ylabel('Y分位数',fontsize=15)
plt.savefig('img/Q1-img/价格转换后的分布.png',dpi=300)
plt.show()
```

结论：newprice和price是属于数值特征，存在异常值，数据分布不符合正态分布，需要对数转换后，便可以近似为正态分布。newprice与price近似线性相关。

# 6 查看相关性

```python
plt.rcParams['font.sans-serif'] = ['STSong']
corr = train.corr().abs()
k = 10
col =  corr.nlargest(k,'价格')['价格'].index
plt.subplots(figsize = (10,10))
sns.heatmap(train[col].corr(),annot=True,square=True,annot_kws={"size":14})
plt.savefig('img/Q1-img/Top10热图.png',dpi=300)
plt.show()
```

```python
# 筛选出相关性大于0.1的特征，作为接下来交叉特征的特征
# sns.set_style('darkgrid', {'font.sans-serif': 'simhei'})
mcorr = train.corr(method='spearman').abs()
tt = mcorr[mcorr['价格'] > 0.1]
plt.figure()
y = tt['价格']
del y['价格']
y.sort_values().plot(kind="barh", color="b", alpha=1,fontsize=12)
plt.savefig(
    'img/Q1-img/筛选出相关性较高的16个特征.png', bbox_inches='tight', dpi=300)
```

结论：newprice、displacement、a_8、a2、maketype、modelyear、a13、mileage、gearbox、a5、cityid、carcode、color、a4、serial、model、brand、a10、transferCount、seatings、a14、a6与价格较为相关，且按相关强度排序
